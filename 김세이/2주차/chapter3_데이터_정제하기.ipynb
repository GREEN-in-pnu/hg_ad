열 삭제하기

```python
import gdown
gdown.download('http://bit.ly/3RhoNho','ns_202104.csv',quiet=False)
```

```python
import pandas as pd
ns_df = pd.read_csv('ns_202104.csv',low_memory=False)
ns_df.head()
```

```python
# 맨 오른쪽의 unnamed:13 열은 각 라인 끝의 콤마를 인식하여 판다스에서 자동으로 추가된 열임. 불필요한 열이므로 삭제해야 함.
# 방법 1 슬라이싱
ns_book = ns_df.loc[:,'번호':'등록일자']
ns_book.head()
```

```python
# 중간의 '부가기호' 열을 제외하고 선택하려면? => 슬라이싱으로는 중간에 있는 열을 제외하기 어려움 => 불리언 배열 사용!
# 데이터프레임 열 이름이 저장된 columns 속성 확인 => 판다스의 index 클래스 객체(원소를 파이썬의 리스트처럼 숫자 인덱스로 참조할 수 있음)
print(ns_df.columns)
# 예를 들어, 다음과 같이 적을 수 있음.
print(ns_df.columns[0])
# 원소별 비교: 판다스 배열 성격의 객체는 어떤 값과 비교할 때 자동으로 배열에 있는 모든 원소와 하나씩 비교해줌.
# 이때 비교 연산 결과를 넘파이 배열로 반환함.
ns_df.columns != 'Unnamed: 13'
```

```python
# 방법 2 불리언 배열 이용
selected_columns = ns_df.columns != 'Unnamed: 13'
ns_book = ns_df.loc[:, selected_columns]
ns_book.head()

# '부가기호' 열 제거하기
selected_columns = ns_df.columns!='부가기호'
ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

```python
# 방법 3 drop() 메서드
# axis=1을 지정하면 열을 삭제함.
ns_book = ns_df.drop('Unnamed: 13', axis=1)
ns_book.head()

# 여러 열 지정
ns_book = ns_df.drop(['부가기호','Unnamed: 13'], axis=1)
ns_book.head()

# 새로운 변수 생성하지 않고 바로 원래 데이터프레임 수정하기 by. inplace 매개변수(선택한 데이터프레임에 덮어쓰기)
ns_book.drop('주제분류번호', axis=1, inplace=True)
ns_book.head()
```

```python
# 방법 4 dropna() 메서드
# NaN이 하나 이상 포함된 행이나 열을 삭제함
ns_book = ns_df.dropna(axis=1)
ns_book.head()

# 모든 값이 NaN인 열을 삭제하려면 how 매개변수 이용
ns_book = ns_df.dropna(axis=1, how='all')
ns_book.head()
```

행 삭제하기

```python
# 방법 1 drop() 메서드 => axis=0(default)
# 직접 행 인덱스를 지정해 삭제하는 방법
ns_book2 = ns_book.drop([0,1])
ns_book2.head()
```

```python
# 방법 2 [] 연산자; 슬라이싱, 불리언 배열 이용
# 슬라이싱
ns_book2 = ns_book[2:]
ns_book2.head()
# 불리언 배열
selected_rows = ns_df['출판사']=='한빛미디어'
ns_book2 = ns_book[selected_rows]
ns_book2.head()

ns_book2 = ns_book[ns_book['대출건수']>1000]
ns_book2.head()
# 실제 사용할 때는 불리언 배열을 만드는 조건을 바로 []연산자에 대입함.
```

중복된 행 찾기

```python
# duplicated() 메서드
sum(ns_book.duplicated())
# 모든 열을 기준으로 중복된 행을 찾는다.

# 일부 열을 기준으로 중복된 행을 찾고싶다면 sunset 매개변수를 활용한다.
sum(ns_book.duplicated(subset=['도서명','저자','ISBN']))
# 총 몇 개의 행이 지정한 열을 기준으로 보았을 때 중복되었는지 알 수 있다.
# keep 매개변수를 false로 지정하면 중복된 행을 모두 true로 표시한 불리언 배열을 반환한다.
dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN'], keep=False)
ns_book3 = ns_book[dup_rows]
ns_book3.head()
```

그룹별로 모으기

```python
# groupby() 메서드
# by 매개변수에 행을 합칠 때 기준이 되는 열을 지정한다.
count_df = ns_book[['도서명','저자','ISBN','권','대출건수']]
group_df = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False)
loan_count = group_df.sum()
loan_count.head()

loan_count = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False).sum()
loan_count.head()
```

원본 데이터 업데이트 하기

```python
# 대출건수를 원본 데이터 프레임에 업데이터하기 위해 거쳐야 할 과정
# 1. duplicated 메서드로 중복된 행을 True로 표시한 불리언 배열을 만든다.
# 2. 1번에서 구한 불리언 배열을 반전시켜(판다스의 ~연산자 사용) 중복되지 않은 고유한 행을 True로 표시한다.
# 3. 2번에서 구한 불리언 배열을 사용해 운본 배열에서 고유한 행만 선택한다.(copy 메서드 사용)
dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN','권'])
unique_rows = ~dup_rows
ns_book3 = ns_book[unique_rows].copy()

sum(ns_book3.duplicated(subset=['도서명','저자','ISBN','권']))
```

```python
# 원본 데이터프레임 인덱스 설정하기
# ns_book3의 인데스를 loan_count 데이터프레임의 인덱스와 동일하게 만들기.
# set_index() 메서드 이용
ns_book3.set_index(['도서명','저자','ISBN','권'],inplace=True)
ns_book3.head()
# loan_count와 동일하게 도서명, 저자, ISBN, 권 열이 인덱스가 되었음을 알 수 있음.
```

```python
# 업데이트하기 => update() 메서드; 다른 데이터프레임을 이용해 원본 데이터프레임의 값을 업데이트함.
ns_book3.update(loan_count)
ns_book3.head()
```

```python
# 인덱스 열 해제 => reset_index() 메서드; 데이터프레임 인덱스 재설정 가능
ns_book4 = ns_book3.reset_index()
ns_book4.head()
```

```python
# 여러 개의 열을 사용해 데이터프레임의 인덱스를 만들었다 다시 해제했으므로 초기 데이터프레임의 열과 ns_book4 데이터프레임의 열 순서가 달라짐.
# 작업 시 혼동을 줄이도록 원래 열의 순서대로 맞추는 작업 진행.
ns_book4 = ns_book4[ns_book.columns]
ns_book4.head()
```

```python
# 불필요한 데이터를 제거하여 만든 ns_book4 데이터프레임을 저장
ns_book4.to_csv('ns_book4.csv',index=False)
```

일괄처리함수: 불필요한 행과 열을 제거하기 위해 작성한 코드를 새로운 데이터에 적용하기 쉽도록 일괄 처리하는 data_cleaning() 함수 만들기.
데이터 정제 과정 단순화.

```python
def data_cleaning(filename):
  ns_df = pd.read_csv(filename, low_memory=False)
  ns_book = ns_df.dropna(axis=1, how='all')
  count_df = ns_book[['도서명','저자','ISBN','권','대출건수']]
  loan_count = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False).sum()
  dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN','권'])
  unique_rows = ~dup_rows
  ns_book3= ns_book[unique_rows].copy()
  ns_book3.set_index(['도서명','저자','ISBN','권'], inplace=True)
  ns_book3.update(loan_count)
  ns_book4 = ns_book3.reset_index()
  ns_book4 = ns_book4[ns_book.columns]
  return ns_book4
```

```python
new_ns_book4 = data_cleaning('ns_202104.csv')
ns_book4.equals(new_ns_book4)
```

3-2. 잘못된 데이터 수정하기
NaN : 판다스에서 누락된 값을 표시하는 기호. isna() 메서드를 이용해 NaN의 여부를 확인하거나 notna() 메서드를 이용해 NaN이 아닌 값인지 체크할 수 있음.
정규표현식 : 문자열에서 패턴을 찾고 대체하기 위한 규칙의 모음. 정규표현식을 사용하면 복잡한 패턴을 가진 문자열을 쉽게 검색할 수 있음.  
| 함수, 메서드 | 기능|  
| ------- | ------ |  
| DataFrame.info() | 데이터프레임의 요약 정보를 출력|  
|isna() | 누락된 값을 감지. 셀의 값이 None, NaN일 경우 True를 반환|  
|astype() | 데이터타입 지정 |   
|fillna() | 누락된 원소의 값을 채움 |  
|replace() | 데이터프레임의 값을 다른 값으로 바꿈 |  
|Series.str.contains() | 시리즈나 인덱스에서 문자열 패턴을 포함하고 있는지 검사|  
|gt() | 데이터프레임의 원소보다 큰 값을 검사 |
